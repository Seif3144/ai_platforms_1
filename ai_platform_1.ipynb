```python
import torch

# Input
x = torch.tensor(1.5, requires_grad=True)
print("Input x:", x.item())
```

```python
# Weights & biases (Layer 1)
w00 = torch.tensor(0.5, requires_grad=True)
b00 = torch.tensor(0.1, requires_grad=True)

w01 = torch.tensor(-0.3, requires_grad=True)
b01 = torch.tensor(0.2, requires_grad=True)

w02 = torch.tensor(0.8, requires_grad=True)
b02 = torch.tensor(-0.1, requires_grad=True)

# Linear
z1_0 = w00 * x + b00
z1_1 = w01 * x + b01
z1_2 = w02 * x + b02

# ReLU
a1_0 = torch.relu(z1_0)
a1_1 = torch.relu(z1_1)
a1_2 = torch.relu(z1_2)

print("\nLayer 1 outputs (ReLU):")
print(a1_0.item(), a1_1.item(), a1_2.item())
```

```python
# Weights & biases (Layer 2)
w10 = torch.tensor(1.0, requires_grad=True)
b10 = torch.tensor(0.0, requires_grad=True)

w11 = torch.tensor(-1.2, requires_grad=True)
b11 = torch.tensor(0.3, requires_grad=True)

# Linear combinations
z2_0 = w10 * a1_0 + b10
z2_1 = w11 * a1_1 + b11

# Sigmoid
a2_0 = torch.sigmoid(z2_0)
a2_1 = torch.sigmoid(z2_1)

print("\nLayer 2 outputs (Sigmoid):")
print(a2_0.item(), a2_1.item())
```

```python
combined = a2_0 + a2_1
a3 = torch.tanh(combined)

print("\nCombined + Tanh output:")
print(a3.item())
```

```python
# Weights & bias (Output layer)
w20 = torch.tensor(0.7, requires_grad=True)
b20 = torch.tensor(0.05, requires_grad=True)

output = w20 * a3 + b20
print("\nFinal output:")
print(output.item())
```

```python
output.backward()

print("\nGradient of output w.r.t input x:")
print(x.grad.item())
```
